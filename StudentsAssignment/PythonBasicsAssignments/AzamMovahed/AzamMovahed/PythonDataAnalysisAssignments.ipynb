```python
#Import Libraries and Load Data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('./datasets/orders.csv')
df=pd.DataFrame(data)
df
```

```python
#Initial Data Exploration
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.describe()
```

```python
#Identify  missing values
df[df['order_date'].isnull()]
```

```python
#find other info about the null value for handle it
print(df[df['customer_id']==2008])
print(df[df['order_id']==1008])
```

```python
#Data Cleaning and Preprocessing
#Convert the order_date column to a datetime format
def parse_date(date):
  try:
    return pd.to_datetime(date,format="%m/%d/%Y")
  except ValueError:
    try:
      return pd.to_datetime(date, format="%Y-%m-%d")
    except ValueError:
      try:
        return pd.to_datetime(date, format="%d %B %Y")
      except ValueError:
        return None
df['parsed-order_date']=df['order_date'].apply(parse_date)
df['parsed-order_date']=pd.to_datetime(df['parsed-order_date'])
df['fill_parsed-order_date']=pd.date_range('2023-05-01', periods=50, freq='D')
df
```

```python
#Identify and handle any missing values
print(df[df['price'].isnull()])
cat_mean=df.groupby('category')['price'].transform('mean')
df['fil_price']=df['price'].fillna(cat_mean)
df.info()
```

```python
#Check for inconsistencies or errors in the data
outlier_mask=(df['price']<0)
outlier_data=df[outlier_mask]
print(outlier_data)
```

```python
df.loc[outlier_mask,'fil_price']=df.groupby('category')['price'].transform('median')
df
```
